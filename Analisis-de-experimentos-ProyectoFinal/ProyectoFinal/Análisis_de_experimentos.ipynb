### Definiendo la función

image.png

```python
#creamos la interfaz par anuestro cliente.
#la funcion va recivier dos parametros x_h que sera de la harina y x_l que s de la leche. estos parametros isabel lo colocara manuealmente al moemento de usar el interfaz

def modelo_receta(x_h, x_l):
  #establecemos los limites en lsoq esta nuestra matriz lo llamremos limte_normalizado
  limite_normalizado = [-1,+1]
  #establecmoes los limtes de los valores que el usuario pondra colocalar en la interfaz, son  los limtess de los valores de la harina y leche
  limite_harina = [0.5, 1.5]
  limite_leche = [0.2, 0.7]

  #con le metodo interp(interpolacion) conseguimos convertir los valores de x_h y x_l a los limites normalazidos de -1 y +1 de nuesta matriz.
  #recive primero una valor numero que sera(x_h), segundo parametro sera el limite cual se encuentra X_h (0.5-1.5 son los limtes de la harina), como tercer parametro tengo indicarle que lo convierta(limite_normlizado)
  x_h_convertido = np.interp(x_h, limite_harina, limite_normalizado)
  x_l_convertido = np.interp(x_l, limite_leche, limite_normalizado)
#creamos o usamos  la frmula de nuestro modelo, no coloamos el error este se ajusta automaticamnte. solo colocmos e parametros
#multiplicamos el parametro de harina por x_h_cn¿onvertido y el parmaetro de leche x_l_convertido
  cantidades = parametros['Intercept'] + parametros['Harina']*x_h_convertido + parametros['Leche']*x_l_convertido
  #usmaos retur para q no devuelva el resultado redondeado
  return round(cantidades)
```

```python
#usamos el modelo, ingreasamo el vlaore de la harina como primer parametro y segundo la leche
#el modelo tiene un 95% de prediccion
modelo_receta(1.0, 0.5)
```
