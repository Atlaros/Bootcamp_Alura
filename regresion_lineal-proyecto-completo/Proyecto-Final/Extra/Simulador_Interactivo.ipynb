## Importando nuestro modelo

```python
import pickle

modelo = open('modelo_precio','rb')
lm_new = pickle.load(modelo)
modelo.close()

area = 30
garaje = 2
banos = 4
chimenea = 4
marmol = 0
pisos = 1

entrada = [[area, garaje, banos, chimenea, marmol, pisos]]

print('{} litros'.format(lm_new.predict(entrada)[0].round(2)))
```

## Ejemplo de un simulador interactivo para Google Colab

https://ipywidgets.readthedocs.io/en/stable/index.html

https://github.com/jupyter-widgets/ipywidgets

```python
# Importando bibliotecas
from ipywidgets import widgets, HBox, VBox
from IPython.display import display

# Creando los controles del formulario
area = widgets.Text(description="Área")
garaje = widgets.Text(description="Garaje")
banos = widgets.Text(description="Baños")
chimenea = widgets.Text(description="Chimenea")
marmol = widgets.Text(description="Mármol")
pisos = widgets.Text(description="Pisos")

boton = widgets.Button(description="Simular")

# Colocando los controles
left = VBox([area, banos, marmol])
right = VBox([garaje, chimenea, pisos])
inputs = HBox([left, right])

# Función de simulación
def simulador(sender):
    entrada=[[
                float(area.value if area.value else 0), #la funcion value la usamos para trasformar los str en  float. ya que los valores colocamos en la caja son str.
                float(garaje.value if garaje.value else 0),
                float(banos.value if banos.value else 0),
                float(chimenea.value if chimenea.value else 0),
                float(marmol.value if marmol.value else 0),
                float(pisos.value if pisos.value else 0)
             ]]
    print('$ {0:.2f}'.format(lm_new.predict(entrada)[0]))

# Asignando la función "simulador" al evento de clic del botón
boton.on_click(simulador)
```

```python
display(inputs, boton)
```
